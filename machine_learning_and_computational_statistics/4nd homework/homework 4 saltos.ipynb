```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

```python
x = np.linspace(1,5,100)
y = 0.5*x+1
plt.plot(x, y, '-r', label='g(x)=0.5x+1')
plt.title('Graph of g(x)')
plt.xlabel('x', color='#1C2833')
plt.ylabel('y', color='#1C2833')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

```python
# Generate data
mean = [2,2]
cov = [[3,2],[2,4]]
N = 50
theta = 0.5

#Define the the model
def f(x,theta):
    return theta*x + 1

X = []
Y = []
for i in range(0, 100):
    x, y = np.random.multivariate_normal(mean, cov, N).T
    
    X.append(x)
    Y.append(y)

X = np.array(X)
Y = np.array(Y)
```

```python
#plot X data
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
#change number from 0 to 99 to see the 100 datasets one by one
number=1
ax.scatter(X[number],Y[number],c='r',marker='o')
ax.set_xlabel('$x_2$')
ax.set_ylabel('$x_1$')
ax.set_zlabel('y')
plt.show()
```

```python
theta_set=[]
for i in range(0, 100):
    X_new = X[i].T
    XTXinv = (np.dot(X_new.T,X_new) ** (-1))
    theta_est = np.dot(XTXinv,X_new.T)
    theta_est = np.dot(theta_est,Y[number])
    theta_set.append(theta_est)
```

```python
fig1 = plt.figure()
ax1 = fig1.add_subplot(111)

for i in range(0, 100):
    x = np.linspace(1,2,10)
    g = theta_set[number]*x+1
    ax1.plot(x, g, '-b')

g = 0.5*x+1
ax1.plot(x, g, '-g', label='g=0.5x+1')
plt.title('Estimates of theta')
plt.xlabel('x', color='#1C2833')
plt.ylabel('g(x)', color='#1C2833')
plt.legend(loc='upper left')
plt.grid()
plt.show()
    
```
